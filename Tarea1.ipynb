```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('https://raw.githubusercontent.com/MinCiencia/Datos-COVID19/master/output/producto5/TotalesNacionales.csv')
```

```python
data
```

```python
vals = list(data.columns)[1:-1]
ids = list(data.columns)[0]
df = pd.melt(data, value_vars=vals, id_vars=ids)
```

```python
df
```

```python
df = df.rename(columns = {'Fecha':'tipo'})
df = df.rename(columns = {'variable':'fecha'})
df = df.rename(columns = {'value':'valor'})
```

```python
m_covid = df[(df['tipo']=='Fallecidos') & (df['valor']>0)]
m_covid = m_covid.reset_index()
m_covid = m_covid.drop(columns =['index'])
m_covid
```

```python
m_covid.plot(x='fecha', y='valor')
```
